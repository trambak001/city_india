```
# ========================================
# INDIAN CITY CLUSTERING ANALYSIS
# Complete Feature-Rich Version for Google Colab
# ========================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
import warnings
import os
from google.colab import files
from IPython.display import display, HTML, Markdown

warnings.filterwarnings('ignore')

# Set beautiful plotting style
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("husl")

# Check Plotly availability
try:
    import plotly.express as px
    import plotly.graph_objects as go
    from plotly.subplots import make_subplots
    PLOTLY_AVAILABLE = True
    print("‚úÖ Plotly available - Interactive visualizations enabled")
except ImportError:
    PLOTLY_AVAILABLE = False
    print("‚ö†Ô∏è  Plotly not available - Using matplotlib only")

# ========================================
# UTILITY FUNCTIONS FOR BEAUTIFUL OUTPUT
# ========================================

def print_banner(text, color='blue'):
    """Print beautiful banner"""
    colors = {
        'blue': '\033[94m',
        'green': '\033[92m',
        'yellow': '\033[93m',
        'red': '\033[91m',
        'end': '\033[0m'
    }
    c = colors.get(color, colors['blue'])
    end = colors['end']
    print(f"\n{c}{'=' * 70}")
    print(f"  {text}")
    print(f"{'=' * 70}{end}\n")

def print_section(text, emoji='üìå'):
    """Print section header"""
    print(f"\n{'‚îÄ' * 70}")
    print(f"{emoji} {text}")
    print(f"{'-' * 70}")

def print_metric(label, value, unit=''):
    """Print formatted metric"""
    print(f"   ‚Ä¢ {label:<30} {value:>15}{unit}")

def display_cluster_card(cluster_id, cluster_title, cities, features_dict, total_cities):
    """Display beautiful cluster information card"""
    print(f"\n{'‚îå' + '‚îÄ' * 68 + '‚îê'}")
    print(f"‚îÇ üåÜÔ∏è  CLUSTER {cluster_id}: {cluster_title:<40} ‚îÇ")
    print(f"‚îÇ {'‚îÄ' * 68} ‚îÇ")
    print(f"‚îÇ Cities: {len(cities)} ({len(cities)/total_cities*100:.1f}% of total){' ' * (68 - 30 - len(str(len(cities))))}‚îÇ")
    print(f"‚îÇ {'‚îÄ' * 68} ‚îÇ")

    # Display cities (max 5 per line)
    city_list = cities.tolist()
    for i in range(0, len(city_list), 5):
        cities_line = ', '.join(city_list[i:i+5])
        if len(cities_line) > 62:
            cities_line = cities_line[:59] + '...'
        print(f"‚îÇ {cities_line:<66} ‚îÇ")

    print(f"‚îÇ {'‚îÄ' * 68} ‚îÇ")
    print(f"‚îÇ üìä Key Characteristics:{' ' * 44} ‚îÇ")

    for feature, value in features_dict.items():
        feature_display = feature[:35]
        value_str = f"{value:.2f}"
        spaces = 68 - len(feature_display) - len(value_str) - 5
        print(f"‚îÇ   ‚Ä¢ {feature_display}:{' ' * spaces}{value_str} ‚îÇ")

    print(f"‚îî{'‚îÄ' * 68}‚îò")

# ========================================
# MAIN ANALYSIS FUNCTION
# ========================================

def analyze_city_clusters(n_clusters_override=None, custom_cluster_labels=None):
    """Complete city clustering analysis with enhanced visualizations"""

    print_banner("üèôÔ∏è  INDIAN CITY CLUSTERING ANALYSIS", 'blue')

    # === FILE UPLOAD ===
    print("‚¨ÜÔ∏è  Please upload your CSV data file:")
    print("   (File should contain: City, Latitude, Longitude, and feature columns)\n")

    uploaded = files.upload()

    if not uploaded:
        print("‚ùå No file uploaded. Analysis cancelled.")
        return None, None

    # Load the uploaded file
    file_name = next(iter(uploaded))
    try:
        df = pd.read_csv(file_name)
        print(f"‚úÖ Data loaded successfully from: {file_name}")
    except Exception as e:
        print(f"‚ùå Error loading file {file_name}: {e}")
        return None, None
    finally:
        if os.path.exists(file_name):
            os.remove(file_name)

    # === DATA OVERVIEW ===
    print_section("üìä Dataset Overview", "üìä")
    print_metric("Total Cities", len(df))
    print_metric("Total Columns", len(df.columns))
    print_metric("Missing Values", df.isnull().sum().sum())
    print_metric("Memory Usage", f"{df.memory_usage(deep=True).sum() / 1024:.2f}", " KB")

    print("\nüìã Column Information:")
    display(pd.DataFrame({
        'Column': df.columns,
        'Type': df.dtypes.values,
        'Non-Null': df.count().values,
        'Missing': df.isnull().sum().values
    }))

    # === DATA PREPROCESSING ===
    print_section("üîß Data Preprocessing", "üîß")

    # Handle missing values
    numeric_columns_for_imputation = df.select_dtypes(include=[np.number]).columns
    missing_filled = 0
    for col in numeric_columns_for_imputation:
        if df[col].isnull().any():
            df[col].fillna(df[col].median(), inplace=True)
            missing_filled += 1

    if missing_filled > 0:
        print(f"‚úÖ Filled missing values in {missing_filled} columns using median")

    # Handle Population column
    if 'Population (millions)' in df.columns:
        df['Population'] = df['Population (millions)'] * 1_000_000
        df = df.drop(columns=['Population (millions)'])
        print("‚úÖ Converted 'Population (millions)' to 'Population'")

    # Dynamically identify the primary identifier column for display purposes and exclusion from features
    identifier_column = None
    if 'City' in df.columns:
        identifier_column = 'City'
    elif 'District name' in df.columns: # Specific check for the current dataset structure
        identifier_column = 'District name'
    else: # Fallback to the first object column if no common identifier is found
        object_cols = df.select_dtypes(include='object').columns
        if len(object_cols) > 0:
            identifier_column = object_cols[0]
            print(f"‚ö†Ô∏è  No 'City' or 'District name' column found. Using '{identifier_column}' as identifier.")
        else:
            print("‚ùå No suitable identifier column found (e.g., 'City' or 'District name'). Analysis may be impacted.")
            # Set a placeholder to avoid errors in display functions, though it won't be meaningful
            identifier_column = 'Unnamed_ID'
            df[identifier_column] = df.index.astype(str)

    # Identify feature columns: all numerical columns, excluding Lat/Lon and the identified identifier column
    feature_columns = df.select_dtypes(include=[np.number]).columns.tolist()

    # Exclude Latitude and Longitude if they are present and numerical (they are not features for clustering)
    if 'Latitude' in feature_columns: feature_columns.remove('Latitude')
    if 'Longitude' in feature_columns: feature_columns.remove('Longitude')

    # Exclude the identifier column if it somehow ended up in numerical features (e.g., if a numeric ID column was present)
    if identifier_column and identifier_column in feature_columns: # Added check for identifier_column existence
        feature_columns.remove(identifier_column)

    if not feature_columns:
        print("‚ùå Error: No numerical features identified for clustering after preprocessing. Cannot proceed.")
        return None, None

    print(f"‚úÖ Identified {len(feature_columns)} features for clustering:")
    for i, col in enumerate(feature_columns, 1):
        print(f"   {i}. {col}")

    # === FEATURE SCALING ===
    X = df[feature_columns].copy()
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    print(f"\n‚úÖ Features scaled using StandardScaler")

    # === OPTIMAL CLUSTER SELECTION ===
    print_section("üéØ Finding Optimal Number of Clusters", "üéØ")

    k_range = range(2, min(11, len(df)))
    inertias = []
    silhouette_scores = []
    davies_bouldin_scores = []
    calinski_scores = []

    print("Computing metrics for k = 2 to 10...")
    for k in k_range:
        kmeans_eval = KMeans(n_clusters=k, random_state=42, n_init=10)
        labels_eval = kmeans_eval.fit_predict(X_scaled)

        inertias.append(kmeans_eval.inertia_)
        silhouette_scores.append(silhouette_score(X_scaled, labels_eval))
        davies_bouldin_scores.append(davies_bouldin_score(X_scaled, labels_eval))
        calinski_scores.append(calinski_harabasz_score(X_scaled, labels_eval))

    # Find optimal k based on metrics (default)
    optimal_k = k_range[np.argmax(silhouette_scores)]

    # Override optimal_k if specified by user
    if n_clusters_override is not None:
        if n_clusters_override < 2 or n_clusters_override > len(df) - 1:
            print(f"‚ùå Invalid n_clusters_override={n_clusters_override}. Must be between 2 and {len(df) - 1}. Using default optimal_k={optimal_k}.")
        else:
            optimal_k = n_clusters_override
            print(f"\n‚úÖ Using user-defined number of clusters: k = {optimal_k}")
            if optimal_k not in k_range:
                # If user-defined k is outside pre-computed range, compute metrics for it
                kmeans_override_eval = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
                labels_override_eval = kmeans_override_eval.fit_predict(X_scaled)
                # Append metrics for overridden k to ensure plots have it if possible
                k_range = list(k_range) + [optimal_k]
                inertias.append(kmeans_override_eval.inertia_)
                silhouette_scores.append(silhouette_score(X_scaled, labels_override_eval))
                davies_bouldin_scores.append(davies_bouldin_score(X_scaled, labels_override_eval))
                calinski_scores.append(calinski_harabasz_score(X_scaled, labels_override_eval))
                # Re-sort lists based on k_range for consistent plotting
                sorted_indices = np.argsort(k_range)
                k_range = [k_range[i] for i in sorted_indices]
                inertias = [inertias[i] for i in sorted_indices]
                silhouette_scores = [silhouette_scores[i] for i in sorted_indices]
                davies_bouldin_scores = [davies_bouldin_scores[i] for i in sorted_indices]
                calinski_scores = [calinski_scores[i] for i in sorted_indices]


    print(f"\n{'Metric':<30} {'Value':<15} {'Interpretation'}")
    print("‚îÄ" * 70)
    if optimal_k in k_range:
        # Ensure we're using the metrics corresponding to the final optimal_k
        idx = k_range.index(optimal_k)
        print_metric("Optimal Clusters (k)", optimal_k)
        print_metric("Silhouette Score", f"{silhouette_scores[idx]:.4f}", " (Higher is better)")
        print_metric("Davies-Bouldin Index", f"{davies_bouldin_scores[idx]:.4f}", " (Lower is better)")
        print_metric("Calinski-Harabasz", f"{calinski_scores[idx]:.2f}", " (Higher is better)")
    else:
        print_metric("Optimal Clusters (k)", optimal_k)
        print("   ‚Ä¢ Metrics for this k were not in the initially computed range.")


    # === CLUSTERING ===
    print_section(f"üî¨ Performing K-Means Clustering (k={optimal_k})", "üî¨")

    kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
    df['Cluster'] = kmeans.fit_predict(X_scaled)

    print(f"‚úÖ Clustering complete!")
    print(f"   Iterations: {kmeans.n_iter_}")
    print(f"   Final inertia: {kmeans.inertia_:.2f}")

    # === CLUSTER NAMING ===
    print_section("üìù Naming Clusters Based on Characteristics", "üìù")

    cluster_profiles_mean = df.groupby('Cluster')[feature_columns].mean()
    cluster_titles = {}

    if custom_cluster_labels is not None and len(custom_cluster_labels) == optimal_k:
        print(f"‚úÖ Using user-defined cluster labels for {optimal_k} clusters.")
        for i in range(optimal_k):
            cluster_titles[i] = custom_cluster_labels[i]
    else:
        print("‚úÖ Using rule-based cluster naming.")
        # Simple rule-based naming (can be customized based on your features)
        for cluster_id in sorted(df['Cluster'].unique()):
            profile = cluster_profiles_mean.loc[cluster_id]
            title = f"Cluster {cluster_id}" # Default title

            # Example naming logic - customize this based on your data and desired titles
            # Add checks for feature existence before using them in comparison
            has_pci = 'Per Capita Income (‚Çπ)' in profile.index
            has_lit = 'Literacy Rate (%)' in profile.index
            has_service = 'Service Sector Employment (%)' in profile.index
            has_industry = 'Number of Industries per 1000 population' in profile.index

            if has_pci and has_lit:
                overall_pci_mean = df['Per Capita Income (‚Çπ)'].mean() if 'Per Capita Income (‚Çπ)' in df.columns else -np.inf
                overall_lit_mean = df['Literacy Rate (%)'].mean() if 'Literacy Rate (%)' in df.columns else -np.inf

                if profile['Per Capita Income (‚Çπ)'] > overall_pci_mean and \
                   profile['Literacy Rate (%)'] > overall_lit_mean:
                    title = "Economically Advanced & Literate"
                elif profile['Per Capita Income (‚Çπ)'] < overall_pci_mean and \
                     profile['Literacy Rate (%)'] < overall_lit_mean:
                     title = "Developing & Less Literate"

            if has_service and title == f"Cluster {cluster_id}": # Only apply if not already named
                overall_service_mean = df['Service Sector Employment (%)'].mean() if 'Service Sector Employment (%)' in df.columns else -np.inf
                if profile['Service Sector Employment (%)'] > overall_service_mean:
                     title = "Service Sector Focused"

            if has_industry and title == f"Cluster {cluster_id}": # Only apply if not already named
                overall_industry_mean = df['Number of Industries per 1000 population'].mean() if 'Number of Industries per 1000 population' in df.columns else -np.inf
                if profile['Number of Industries per 1000 population'] > overall_industry_mean:
                     title = "Industrially Focused"


            cluster_titles[cluster_id] = title


    for cluster_id in sorted(df['Cluster'].unique()):
        print(f"Cluster {cluster_id}: {cluster_titles[cluster_id]}")


    # === VISUALIZATIONS ===
    print_section("üìä Generating Visualizations", "üìä")

    # 1. Elbow Method and Silhouette Analysis
    print("üìà Creating optimization plots...")
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))

    # Elbow Method
    axes[0, 0].plot(k_range, inertias, 'bo-', linewidth=2, markersize=8)
    axes[0, 0].set_title('Elbow Method for Optimal k', fontsize=14, fontweight='bold')
    axes[0, 0].set_xlabel('Number of Clusters (k)', fontsize=11)
    axes[0, 0].set_ylabel('Inertia', fontsize=11)
    axes[0, 0].grid(True, alpha=0.3)
    axes[0, 0].axvline(x=optimal_k, color='red', linestyle='--', alpha=0.7, label=f'Chosen k={optimal_k}')
    axes[0, 0].legend()

    # Silhouette Score
    axes[0, 1].plot(k_range, silhouette_scores, 'go-', linewidth=2, markersize=8)
    axes[0, 1].set_title('Silhouette Score Analysis', fontsize=14, fontweight='bold')
    axes[0, 1].set_xlabel('Number of Clusters (k)', fontsize=11)
    axes[0, 1].set_ylabel('Silhouette Score', fontsize=11)
    axes[0, 1].grid(True, alpha=0.3)
    axes[0, 1].axvline(x=optimal_k, color='red', linestyle='--', alpha=0.7, label=f'Chosen k={optimal_k}')
    axes[0, 1].legend()

    # Davies-Bouldin Index
    axes[1, 0].plot(k_range, davies_bouldin_scores, 'ro-', linewidth=2, markersize=8)
    axes[1, 0].set_title('Davies-Bouldin Index (Lower is Better)', fontsize=14, fontweight='bold')
    axes[1, 0].set_xlabel('Number of Clusters (k)', fontsize=11)
    axes[1, 0].set_ylabel('Davies-Bouldin Index', fontsize=11)
    axes[1, 0].grid(True, alpha=0.3)
    axes[1, 0].axvline(x=optimal_k, color='red', linestyle='--', alpha=0.7, label=f'Chosen k={optimal_k}')
    axes[1, 0].legend()

    # Calinski-Harabasz Score
    axes[1, 1].plot(k_range, calinski_scores, 'mo-', linewidth=2, markersize=8)
    axes[1, 1].set_title('Calinski-Harabasz Score (Higher is Better)', fontsize=14, fontweight='bold')
    axes[1, 1].set_xlabel('Number of Clusters (k)', fontsize=11)
    axes[1, 1].set_ylabel('Calinski-Harabasz Score', fontsize=11)
    axes[1, 1].grid(True, alpha=0.3)
    axes[1, 1].axvline(x=optimal_k, color='red', linestyle='--', alpha=0.7, label=f'Chosen k={optimal_k}')
    axes[1, 1].legend()

    plt.tight_layout()
    plt.savefig('optimization_plots.png', dpi=300) # Save figure
    plt.show()


    # 2. Geographic Visualization (Matplotlib)
    if 'Latitude' in df.columns and 'Longitude' in df.columns and 'Population' in df.columns: # Added checks for Latitude/Longitude presence
        print("üó∫Ô∏è  Creating geographic visualization...")

        fig, ax = plt.subplots(figsize=(18, 12))
        colors = plt.cm.Set3(np.linspace(0, 1, optimal_k)) # Keep the original colors for Matplotlib

        for i in range(optimal_k):
            cluster_data = df[df['Cluster'] == i]
            scatter = ax.scatter(cluster_data['Longitude'], cluster_data['Latitude'],
                                c=[colors[i]], s=cluster_data['Population']/40000,
                                alpha=0.7, label=f'Cluster {i}: {cluster_titles[i]}', edgecolors='black', linewidth=1)

        ax.set_title('Indian Cities Clustering - Geographic Distribution',
                    fontsize=18, fontweight='bold', pad=20)
        ax.set_xlabel('Longitude', fontsize=13)
        ax.set_ylabel('Latitude', fontsize=13)
        ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=11)
        ax.grid(True, alpha=0.3)

        # Add city labels for top 15 cities
        top_cities = df.nlargest(15, 'Population')
        for _, city in top_cities.iterrows():
            ax.annotate(city[identifier_column], (city['Longitude'], city['Latitude']), # Changed 'City' to identifier_column
                       xytext=(5, 5), textcoords='offset points',
                       fontsize=9, alpha=0.8,
                       bbox=dict(boxstyle='round,pad=0.3', facecolor='yellow', alpha=0.3))

        plt.tight_layout()
        plt.savefig('geographic_map.png', dpi=300) # Save figure
        plt.show()
    else:
        print("‚ö†Ô∏è  Skipping geographic visualization: 'Latitude' and/or 'Longitude' columns not found in data.")

    # 3. Cluster Size Distribution
    print("üìä Creating cluster distribution plots...")
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))

    # Cluster sizes
    cluster_sizes = df['Cluster'].value_counts().sort_index()
    axes[0].bar(cluster_sizes.index, cluster_sizes.values,
               color=plt.cm.Set3(np.linspace(0, 1, optimal_k)),
               edgecolor='black', linewidth=1.5)
    axes[0].set_title('Cities per Cluster', fontsize=14, fontweight='bold')
    axes[0].set_xlabel('Cluster', fontsize=11)
    axes[0].set_ylabel('Number of Cities', fontsize=11)
    axes[0].set_xticks(cluster_sizes.index)
    axes[0].set_xticklabels([f'{i}\n({cluster_titles[i]})' for i in cluster_sizes.index])
    axes[0].grid(True, alpha=0.3, axis='y')

    # Add value labels on bars
    for i, v in enumerate(cluster_sizes.values):
        axes[0].text(i, v + 0.5, str(v), ha='center', va='bottom', fontweight='bold')

    # Cluster percentages
    cluster_pcts = (cluster_sizes / len(df) * 100).values
    axes[1].pie(cluster_pcts, labels=[f'Cluster {i}\n({cluster_titles[i]})' for i in cluster_sizes.index],
               autopct='%1.1f%%', startangle=90,
               colors=plt.cm.Set3(np.linspace(0, 1, optimal_k)),
               textprops={'fontsize': 11, 'fontweight': 'bold'})
    axes[1].set_title('Cluster Distribution (%)', fontsize=14, fontweight='bold')

    plt.tight_layout()
    plt.savefig('cluster_distribution_plots.png', dpi=300) # Save figure
    plt.show()

    # 4. Feature Analysis Heatmap
    print("üî• Creating cluster characteristics heatmap...")
    plt.figure(figsize=(14, 10))

    cluster_features = df.groupby('Cluster')[feature_columns].mean()

    # Normalize for better visualization
    cluster_features_norm = (cluster_features - cluster_features.mean()) / cluster_features.std()

    sns.heatmap(cluster_features_norm.T, annot=True, fmt='.2f',
               cmap='RdYlGn', center=0, linewidths=1, linecolor='white',
               cbar_kws={'label': 'Normalized Values (Z-score)'})
    plt.title('Cluster Characteristics Heatmap\n(Normalized Feature Values)',
             fontsize=16, fontweight='bold', pad=15)
    plt.xlabel('Cluster', fontsize=12, fontweight='bold')
    plt.ylabel('Features', fontsize=12, fontweight='bold')
    plt.xticks(ticks=np.arange(optimal_k) + 0.5, labels=[cluster_titles[i] for i in sorted(df['Cluster'].unique())], rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.savefig('cluster_heatmap.png', dpi=300) # Save figure
    plt.show()

    # 5. Feature Comparison Box Plots
    print("üì¶ Creating feature comparison box plots...")
    n_features = len(feature_columns)
    n_cols = 3
    n_rows = (n_features + n_cols - 1) // n_cols

    fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, 5*n_rows))
    axes = axes.flatten() if n_features > 1 else [axes]

    for idx, feature in enumerate(feature_columns):
        sns.boxplot(data=df, x='Cluster', y=feature, ax=axes[idx],
                   palette='Set3')
        axes[idx].set_title(f'{feature} by Cluster', fontsize=11, fontweight='bold')
        axes[idx].set_xlabel('Cluster', fontsize=10)
        axes[idx].set_ylabel(feature, fontsize=10)
        axes[idx].set_xticks(sorted(df['Cluster'].unique()))
        axes[idx].set_xticklabels([cluster_titles[i] for i in sorted(df['Cluster'].unique())], rotation=45, ha='right')
        axes[idx].grid(True, alpha=0.3, axis='y')


    # Hide unused subplots
    for idx in range(n_features, len(axes)):
        axes[idx].set_visible(False)

    plt.tight_layout()
    plt.savefig('feature_box_plots.png', dpi=300) # Save figure
    plt.show()


    # 6. Interactive Plotly Visualizations
    if PLOTLY_AVAILABLE and 'Population' in df.columns and 'Latitude' in df.columns and 'Longitude' in df.columns: # Added checks for Latitude/Longitude presence
        print("üåç Creating interactive map...")

        # Create hover data dynamically
        hover_data_dict = {col: True for col in feature_columns[:5]}  # Show first 5 features
        hover_data_dict['Cluster Title'] = df['Cluster'].map(cluster_titles)


        fig = px.scatter_mapbox(df, lat="Latitude", lon="Longitude",
                               color=df['Cluster'].map(cluster_titles), size="Population",
                               hover_name=identifier_column, # Changed 'City' to identifier_column
                               hover_data=hover_data_dict,
                               mapbox_style="open-street-map",
                               title="Interactive Indian Cities Clustering Map",
                               zoom=4, center={"lat": 20, "lon": 78},
                               color_discrete_sequence=px.colors.qualitative.T10) # Changed color sequence


        fig.update_layout(height=700, width=1200, font=dict(size=12))
        fig.show()
    else:
        if PLOTLY_AVAILABLE:
            print("‚ö†Ô∏è  Skipping interactive map visualization: 'Latitude' and/or 'Longitude' columns not found in data.")
        else:
            print("‚ö†Ô∏è  Plotly not available for interactive map visualization.")

    # Feature correlation matrix
    print("üîó Creating feature correlation matrix...")
    feature_corr = df[feature_columns].corr()
    fig_corr = px.imshow(feature_corr, text_auto='.2f', aspect="auto",
                            title="Feature Correlation Matrix",
                            color_continuous_scale='RdBu_r',
                            color_continuous_midpoint=0)
    fig_corr.update_layout(height=600, width=800)
    fig_corr.show()

    # 3D Scatter (if enough features)
    if len(feature_columns) >= 3:
        print("üé≤ Creating 3D cluster visualization...")
        fig_3d = px.scatter_3d(df, x=feature_columns[0], y=feature_columns[1],
                                  z=feature_columns[2], color=df['Cluster'].map(cluster_titles),
                                  hover_name=identifier_column, title='3D Cluster Visualization', # Changed 'City' to identifier_column
                                  color_discrete_sequence=px.colors.qualitative.T10) # Changed color sequence
        fig_3d.update_layout(height=700, width=1000)
        fig_3d.show()


    # === DETAILED CLUSTER ANALYSIS ===
    print_section("üìã Detailed Cluster Analysis Results", "üìã")

    total_cities = len(df)

    for cluster_id in sorted(df['Cluster'].unique()):
        cluster_data = df[df['Cluster'] == cluster_id]

        # Prepare feature statistics
        features_dict = {}
        for feature in feature_columns[:8]:  # Show top 8 features
            if feature in cluster_data.columns:
                features_dict[feature] = cluster_data[feature].mean()

        if 'Population' in cluster_data.columns:
            features_dict['Total Population'] = cluster_data['Population'].sum()

        display_cluster_card(cluster_id, cluster_titles[cluster_id], cluster_data[identifier_column], features_dict, total_cities) # Changed 'City' to identifier_column

    # === SUMMARY STATISTICS ===
    print_section("üìä Summary Statistics Table", "üìä")

    summary_stats = df.groupby('Cluster').agg({
        identifier_column: 'count', # Changed 'City' to identifier_column
        **{col: ['mean', 'std', 'min', 'max'] for col in feature_columns}
    }).round(2)

    # Add cluster titles to summary table index
    summary_stats.index = [f"{i}: {cluster_titles[i]}" for i in summary_stats.index]

    display(summary_stats)

    # === INSIGHTS GENERATION ===
    print_section("üí° Business Insights & Recommendations", "üí°")

    insights = []

    # Find best performing cluster for each feature
    for feature in feature_columns[:5]:  # Top 5 features
        if feature in df.columns:
            best_cluster_id = df.groupby('Cluster')[feature].mean().idxmax()
            best_cluster_title = cluster_titles[best_cluster_id]
            best_value = df.groupby('Cluster')[feature].mean().max()
            insights.append(f"üèÜ {best_cluster_title} leads in {feature} (avg: {best_value:.2f})")

    # Population insights
    if 'Population' in df.columns:
        largest_cluster_id = df.groupby('Cluster')['Population'].sum().idxmax()
        largest_cluster_title = cluster_titles[largest_cluster_id]
        insights.append(f"üë• {largest_cluster_title} has the highest total population")

    # Balanced cluster
    cluster_sizes = df['Cluster'].value_counts()
    most_balanced_id = cluster_sizes.idxmax()
    most_balanced_title = cluster_titles[most_balanced_id]
    insights.append(f"‚öñÔ∏è  {most_balanced_title} is the largest with {cluster_sizes.max()} cities")

    for i, insight in enumerate(insights, 1):
        print(f"{i}. {insight}")

    # === EXPORT RESULTS ===
    print_section("üíæ Exporting Results", "üíæ")

    # Save clustered data
    output_file = 'cities_with_clusters_enhanced.csv'
    df['Cluster Title'] = df['Cluster'].map(cluster_titles) # Add cluster title column
    df.to_csv(output_file, index=False)
    print(f"‚úÖ Saved: {output_file}")

    # Save cluster summary
    summary_file = 'cluster_analysis_summary.csv'
    summary_stats.to_csv(summary_file)
    print(f"‚úÖ Saved: {summary_file}")

    # Save cluster profiles
    cluster_profiles = df.groupby('Cluster')[feature_columns].mean().round(2)
    cluster_profiles.index = [cluster_titles[i] for i in cluster_profiles.index] # Add cluster titles to index
    cluster_profiles.to_csv('cluster_profiles.csv')
    print(f"‚úÖ Saved: cluster_profiles.csv")


    # Create downloadable report
    print("\nüì• Download your results:")
    files.download(output_file)
    files.download(summary_file)
    files.download('cluster_profiles.csv')

    # Also make the saved images available for download
    image_files = ['optimization_plots.png', 'geographic_map.png',
                   'cluster_distribution_plots.png', 'cluster_heatmap.png',
                   'feature_box_plots.png']
    print("\nüì• Download plot images:")
    for img_file in image_files:
        if os.path.exists(img_file):
            files.download(img_file)
            print(f"‚úÖ Saved: {img_file}")


    print_banner("üéâ Analysis Complete! Ready for Decision Making üöÄ", 'green')

    return df, summary

# ========================================
# RUN THE ANALYSIS
# ========================================

if __name__ == "__main__":
    # Define custom labels
    custom_labels = [
        "Tier 1: Metro Hubs",
        "Tier 2: Industrial Corridors",
        "Tier 3: Emerging Cities",
        "Tier 4: Developing Towns"
    ]
    # Call the analysis function with override and custom labels
    df_result, summary = analyze_city_clusters(n_clusters_override=4, custom_cluster_labels=custom_labels)
```
